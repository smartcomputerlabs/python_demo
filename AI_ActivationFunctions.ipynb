# Types of activation Functions

1. linear Activation Functions:
   - Function - f(x) = ax
   - Used - Regression models

2. Sigmoid and Logistic Activation Functions
    - function
    - Used in binary Classification

3. Tanh (Hyperbolic Tangent) Activation function
    - Used in hidden layers of a neural network

4. ReLU - (Rectified Linear Unit) Activation Function
    - Function = f(x) = max(0, x)
    - Used - Used for the neural networks, esp Deep neural networks
5. Leaky ReLU
    - Function = f(x) = max(0.01x, x)
    - Used to activate the "Dying neurons" in ReLU

6. Softmax
    - Converts the out puts scores - (logits)



## Choosing an activation Function
- ReLU - Is good for hidden layers due to its simplicity and efficiency
- Sigmoid / SOftwax - Good where the out tasks need probability output
- Tanh - Good for tasks having hidden layers where data needs to be normalized
- Leaky ReLU - If you are having dying neurons

```python
# import the neccesary libraries
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the activation function
sigmoid = nn.Sigmoid()
tanh = nn.Tanh()
relu = nn.ReLU()
leaky_relu = nn.LeakyReLU(0.01)
softmax =nn.Softmax(dim=0)
```

```python
# Generate some data
inputs = torch.linspace(-10, 10, 100)
```

```python
# Apply the activation functions to each of the input data we have
outputs_sigmoid = sigmoid(inputs)
outputs_tanh = tanh(inputs)
outputs_relu = relu(inputs)
outputs_leakey_relu = leaky_relu(inputs)
outputs_softmax = softmax(torch.exp(inputs))
```

```python
# Plots of the outputs
plt.figure(figsize=(12,8))

plt.subplot(2,3,1)
plt.plot(inputs.numpy(), outputs_sigmoid.numpy(), label='Sigmoid')
plt.title('Sigmoid Activation')
plt.xlabel('Input')
plt.ylabel('Output')

plt.subplot(2,3,2)
plt.plot(inputs.numpy(), outputs_tanh.numpy(), label='Tanh')
plt.title('Tanh Activation')
plt.xlabel('Input')
plt.ylabel('Output')

plt.subplot(2,3,3)
plt.plot(inputs.numpy(), outputs_relu.numpy(), label="ReLU")
plt.title('ReLU Activation')
plt.xlabel('Input')
plt.ylabel('Output')

plt.subplot(2,3,4)
plt.plot(inputs.numpy(), outputs_leakey_relu.numpy(), label = 'Leaky ReLU')
plt.title('Leakey ReLU Activation')
plt.xlabel('Input')
plt.ylabel('Output')

plt.subplot(2,3,5)
plt.plot(inputs.numpy(), outputs_softmax.numpy(), label = 'Softmax')
plt.title('SOftmax Activation')
plt.xlabel('Input')
plt.ylabel('Output')

plt.tight_layout()
plt.show()
```
